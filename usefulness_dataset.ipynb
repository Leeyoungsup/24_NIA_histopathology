```python
from glob import glob
import os
import sys
import pandas as pd
import json
from tqdm import tqdm
from random import shuffle
stomach=['위염','미만형선암','장형선암','혼합형선암']
breast=['정상 유방조직','침윤암','상피내암']
```

```python
def create_dir(dir):
    if not os.path.exists(dir):
        os.makedirs(dir)
def split_into_three(total,count):
    # 각 숫자의 몫과 나머지 계산
    quotient, remainder = divmod(total, count)
    # 나머지를 이용해 리스트 생성
    result = [quotient + 1 if i < remainder else quotient for i in range(count)]
    return result

```

```python
total = 100
source_list=[]
label_list=[]
image_list=glob(f'../../data/synth/010.위암 병리 이미지 및 판독문 합성 데이터/1.데이터/**/1.원천데이터/{stomach[0]}/*.png', recursive=True)
shuffle(image_list)
for j in range(50):
    source_list.append(image_list[j])
    label_list.append(0)
total-=50
numbers = split_into_three(total,len(stomach[1:]))
for i in range(len(stomach[1:])):
    image_list=glob('../../data/synth/010.위암 병리 이미지 및 판독문 합성 데이터/1.데이터/**/1.원천데이터/'+stomach[i+1]+'/*.png', recursive=True)
    shuffle(image_list)
    for j in range(numbers[i]):
        source_list.append(image_list[j])
        label_list.append(1)
df=pd.DataFrame({'image':source_list,'label':label_list})
create_dir('../../data/usefulness/stomach/')
df.to_csv('../../data/usefulness/stomach/source.csv',index=False,encoding='utf-8-sig')
```

```python
total = 100
raw_list=[]
label_list=[]
test_list=[]
test_label_list=[]
NT_image_list=glob(f'../../data/origin_type/{stomach[0]}/**/*.jpeg', recursive=True)
shuffle(NT_image_list)
for j in range(50):
    raw_list.append(NT_image_list[j])
    label_list.append(0)
for j in range(50,250):
    test_list.append(NT_image_list[j])
    test_label_list.append(0)
total-=50
numbers = split_into_three(total,len(stomach[1:]))
test_numbers = split_into_three(200,len(stomach[1:]))
for i in range(len(stomach[1:])):
    image_list=glob('../../data/origin_type/'+stomach[i+1]+'/**/*.jpeg', recursive=True)
    shuffle(image_list)
    for j in range(numbers[i]):
        raw_list.append(image_list[j])
        label_list.append(1)
    for j in range(test_numbers[i]):
        test_list.append(image_list[j+numbers[i]])
        test_label_list.append(1)
df=pd.DataFrame({'image':raw_list,'label':label_list})
create_dir('../../data/usefulness/stomach/')
df.to_csv('../../data/usefulness/stomach/raw.csv',index=False,encoding='utf-8-sig')

for i in range(20):
    image_list=[]
    label_list=[]
    for j in range(10):
        image_list.append(test_list[i*10+j])
        image_list.append(test_list[200+i*10+j])
        label_list.append(test_label_list[i*10+j])
        label_list.append(test_label_list[200+i*10+j])
    df=pd.DataFrame({'image':image_list,'label':label_list})
    create_dir('../../data/usefulness/stomach/')
    df.to_csv(f'../../data/usefulness/stomach/test_{i}.csv',index=False,encoding='utf-8-sig')
    
    
```
